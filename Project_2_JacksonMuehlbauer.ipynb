**What is the Target?**

I am interested in completing a classification type problem, thus, there are 2 options I could see as the target for this dataset: 
1. 'code', this feature represents the outcome of the pitch. For example, 'S' is a swinging strike.
2. 'pitch_type', this feature is the category that the pitch falls into. For example, 'FF' is four-seam fastball. 

If I were to move forward with this problem, I would want to build a model that predicts the 'pitch_type'. The purpose of this model would simply be to classify and display the predicted pitch type to a TV or live audience such that they could follow the game more closely. The reason I chose 'pitch_type' over 'code' is to significantly simplify the model... more to come.

**What does one row represent?**
One row represents a single thrown pitch in an MLB game.

**Is this a classification or regression problem?**

As stated earlier, I would treat this as a classification problem. 

**How many features does the data have?**

This dataset has 40 features (including the potential target). While this is a lot, by choosing to classify 'pitch_type', the following features can be dropped as they will have little to no influence on classifying what type of pitch was thrown (or they are empty):

['spin_rate', 'spin_dir', 'sz_bot', 'sz_top', 'type_confidence', 'nasty', 'zone', 'code', 'type', 'event_num', 'b_score', 'ab_id', 'b_count', 's_count', 'outs', 'pitch_num', 'on_1b', 'on_2b', 'on_3b', 'y0'] 

This brings the total features down to 19 with 1 target. 

**How many rows are in the dataset?**

There are 728790 rows in this data set. That said, a much small subset can easily be used for this type of problem

**What, if any, challenges do you foresee in cleaning, exploring, or modeling with this dataset?**

There are many features and many rows in this dataset (with some null values). It could require a fair amount of effort to clean this set. Additionally, the large number of rows and columns could increase the time for training and could potentially disqualify lazy models such as KNN. The last challenge is that there is no "positive" class, this is a multiclass problem, were each class is no more important than the other. This may make evaluating the model metrics harder to interpret.
